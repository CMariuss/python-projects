```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
solar_yield = pd.read_csv('dataset/Solar_Generation_Data.csv')

date_time_1 = solar_yield['DATE_TIME']
np_array = np.array(date_time_1)
```

```python
hour_minute = pd.to_datetime(np_array, format='%d-%m-%Y %H:%M')

solar_yield['DATE_TIME'] = hour_minute

print(solar_yield['DATE_TIME'])
```

```python
randament_energie_solara = solar_yield.where((solar_yield['DATE_TIME'].dt.day == 25) & (solar_yield['DATE_TIME'].dt.month == 5) & (solar_yield['DATE_TIME'].dt.hour > 6) & (solar_yield['DATE_TIME'].dt.hour < 10))
print(randament_energie_solara)
```

```python
print(np.sum(randament_energie_solara['DC_POWER']))
```

```python
# setam graficul pentru randamentul de energie intre orele 7:00 - 10:00
point_plot = sns.lineplot(x='DATE_TIME', y='DC_POWER', data=randament_energie_solara)

# setam titlul graficului
point_plot.set_title('Solar Energy between 7:00-10:00 from form 25-05-2020')
```
