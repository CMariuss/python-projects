Import all necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

Save the data from 'Solar_Generation_Data.csv' database, in a DataFrame

```python
solar_yield = pd.read_csv('dataset/Solar_Generation_Data.csv')
```

Extract the 'DATE_TIME' column, and rewrite it as a DataFrame date time format

```python
solar_yield['DATE_TIME'] = pd.to_datetime(solar_yield['DATE_TIME'])
```

Add a new column called 'TIME' that will contain only the hours and minutes from the 'DATE_TIME' column

```python
solar_yield['TIME'] = solar_yield['DATE_TIME'].dt.strftime('%H:%M')
```

Print the 'TIME' column from DataFrame, to ensure that it contains only the hours and minutes

```python
print(solar_yield['TIME'])
```

Extract the data from May 16 between 6 and 9 AM

```python
specific_date_time_yield = solar_yield.where((solar_yield['DATE_TIME'].dt.day == 16) & (solar_yield['DATE_TIME'].dt.month == 5) & (solar_yield['DATE_TIME'].dt.hour > 5) & (solar_yield['DATE_TIME'].dt.hour < 10))
```

Print the total solar yeld from May 16 between 6 and 9 AM

```python
print(np.sum(specific_date_time_yield['DAILY_YIELD']))
```

Setup a Seaborn diagram to visualise the results

```python
plt.figure(figsize=(12, 6))

# Setup the diagram with the informations about solar yeld between 6 and 9 AM from May 16
point_plot = sns.pointplot(x='TIME', y='DAILY_YIELD', data=specific_date_time_yield)

# Set the diagram title
point_plot.set_title('Solar yield between 6 and 10 AM for May 16')
```
