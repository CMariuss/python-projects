Import all necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

Save all data in a DataFrame

```python
solar_yield = pd.read_csv('dataset/Solar_Generation_Data.csv')
```

Extract the 'DATE_TIME' column from DataFrame and save it in a numpy array

```python
date_time = solar_yield['DATE_TIME']
np_array = np.array(date_time)
```

Print the 'DATE_TIME' column from DataFrame

```python
print(solar_yield['DATE_TIME'])
```

Save the date time from numpy array in a special Pandas datetime format, and replace the original column from DataFrame with it

```python
hour_minute = pd.to_datetime(np_array, format='%d-%m-%Y %H:%M')

solar_yield['DATE_TIME'] = hour_minute
```

Print the 'DATE_TIME' column, to see if the it contain the date time in a desired format

```python
print(solar_yield['DATE_TIME'])
```

Extract the data from May 16 between 6 and 9 AM

```python
specific_date_time_yield = solar_yield.where((solar_yield['DATE_TIME'].dt.day == 16) & (solar_yield['DATE_TIME'].dt.month == 5) & (solar_yield['DATE_TIME'].dt.hour > 5) & (solar_yield['DATE_TIME'].dt.hour < 9))
```

Print the total solar yeld from May 16 between 6 and 9 AM

```python
print(np.sum(specific_date_time_yield['DAILY_YIELD']))
```

```python
# print(len(specific_date_time_yield['DATE_TIME']))
print(specific_date_time_yield['DAILY_YIELD'][100:1000])
```

Initialize the diagram general informations

```python
# Setup the diagram with the informations about solar yeld between 6 and 9 AM from May 16
point_plot = sns.pointplot(x='DATE_TIME', y='DAILY_YIELD', data=specific_date_time_yield)

# Set the diagram title
point_plot.set_title('Solar yield between 6 and 9 AM from May 16')

# Set the hour labels
point_plot.set_xticklabels(['6:00', '6:15', '6:30', '6:45', '7:00', '7:15', '7:30', '7:45', '8:00', '8:15', '8:30', '8:45'])
```
